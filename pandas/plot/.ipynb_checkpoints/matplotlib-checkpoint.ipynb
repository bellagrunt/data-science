# Matplotlib

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
# functional
```

```python
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# OBJECT O
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('Large plat')

axes2.plot(y,x)
axes2.set_title('Small plot')
```

```python
fig,axes = plt.subplots(nrows=3,ncols=3)

# axes.plot(x,y)
plt.tight_layout()
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

plt.tight_layout()

for current_ax in axes:
    current_ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y)
axes[0].set_title('First')

axes[1].plot(y,x)
axes[1].set_title('Second')
```

```python
# Figure size and DPI
```

```python
fig = plt.figure(figsize=(5,2), dpi=100)

ax = fig.add_axes([0,0,1,1])                                         
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1, figsize=(5,4))

axes[0].plot(x,y)

axes[1].plot(y,x)

plt.tight_layout()
```
